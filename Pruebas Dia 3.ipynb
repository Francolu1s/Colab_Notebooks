Bausi ha destinado un presupuesto de $ 15’000.000 para pago de obligaciones a 12 proveedores.
Desarrolle un programa que pida una cantidad a pagar (cuota) para cada 
proveedor y deduzca del presupuesto esas cantidades.
Si el total de las cuotas pagadas supera el presupuesto en algún momento, debe detenerse el ingreso de cuotas y mostrarse el valor total de lo pagado y el valor que no pudo pagarse al excederse el presupuesto de los $15’000.000.
En caso contrario, mostrar el valor total de lo pagado y el presupuesto sobrante.

```python
Presupuesto = 15000000; cuota = 1; ExcedenteFaltante =0; Disponible=0; Pagado=0;

while Pagado <= Presupuesto and cuota < 13:
  valorCuota = int(input(f"Cuota {cuota}:  $"))
  Pagado = Pagado + valorCuota
  cuota += 1

if Pagado > Presupuesto:
  print(f"Total Pagado:  ${Presupuesto}")
  ExcedenteFaltante = Pagado - Presupuesto
  print(f"Excedente que no pudo pagarse:  ${ExcedenteFaltante}")
else:  
  print(f"Total Pagado:  ${Pagado}")
  Disponible = Presupuesto - Pagado
  print(f"Dinero disponible del presupuesto: ${Disponible}")
```
