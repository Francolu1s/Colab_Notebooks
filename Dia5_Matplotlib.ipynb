1. Imports

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.polynomial import Polynomial
import pandas as pd
```

2. Escribimos Funciones

```python
def poly(coeficientes):
  return lambda x: sum([coef*x**i for i, coef in enumerate(coeficientes)])
#log2(x)
f = lambda X: np.log2(x)
#1+x+2*x**2
g=lambda x: 1+x+2*x**2
#Cualquier polinomica 
h=poly([1,2,1]) #1+2x*x+x**2
h2= Polynomial([1,2,1])

h(1999)==h2(1999)

xdis= np.arange(0.01,2,0.2)
x= np.linspace(0.01,2,200)

h2.roots()
h2.deriv()
```

3. Graficamos

```python
plt.title("Grafica de funciones a Prueba")
plt.grid()
plt.axvline(x=0, color="r")
plt.axhline(y=0, color="r")
plt.plot(x,f(x))
plt.plot(xdis,g(xdis),"v")
plt.plot(x,h2(x))
```
