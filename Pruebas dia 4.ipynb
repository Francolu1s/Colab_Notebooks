```python
valores = [50, 30, 80, 45, 20, 10, 15, 15, 30, 50, 12, 100, 12]
contador = 0;
encontrado = False; Sumedia = 0;

valorBuscar = int(input("Digite el valor a buscar en la lista: "))

for valor in valores:
  Sumedia += valor

  if valor == valorBuscar:
#print("Encontrado")
    contador +=1
    encontrado = True  

if encontrado == False:
  print(f"Este elemento '{valorBuscar}' no fue encontrado ")
else:
  print(f"{valorBuscar} fue encontrado {contador} veces")

print(f"el Promedio de los valores es: {Sumedia/len(valores)}")
```

```python
subject = ["Matematicas", "Fisica", "Quimica", "Historia", "Lengua"]
aprobadas = [];

for Materia in subject:
  Nota =float(input(f"Por favor ingrese la nota obtenida en {Materia}: "))
  
  if Nota >=7:
    aprobadas.append(Materia)
    print(Materia)


for Materia in aprobadas:
  print(Materia)
  subject.remove(Materia)

print(f"Asignaturas por repetir: {subject}")
```
